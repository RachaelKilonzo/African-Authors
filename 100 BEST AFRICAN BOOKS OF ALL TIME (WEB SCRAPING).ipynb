```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.shortform.com/best-books/genre/best-african-books-of-all-time'
```

```python
page=requests.get(url)
```

```python
page.encoding='utf-8'
```

```python
soup=BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('div', class_='col-9 col-md-12')
```

```python
soup.find_all('span', class_='d-block d-md-inline')
```

```python
authors=soup.find_all('span', class_='d-block d-md-inline')
```

```python
African_Authors=[author.text for author in authors]
print(African_Authors)
```

```python
import pandas as pd
df = pd.DataFrame({'Author': African_Authors})
df
```

```python
books=soup.find_all('h2', class_='display-4 mb-0 mt--2')
books
```

```python
Books=[book.text for book in books]
print(Books)
```

```python
df['Book'] = Books
df
```

```python
##RATINGS
```

```python
ratings=soup.find_all('p', class_='justify-content-center byline pt-1')
print(ratings)
```

```python
Ratings=[rating.text for rating in ratings]
print(Ratings)
```

```python
Ratings = [float(entry.split('|')[1].strip()) for entry in Ratings]

print(Ratings)
```

```python
df['Rating']=Ratings
df
```
