```python
import requests

def get_book_details(book_title):
    url = f"https://www.googleapis.com/books/v1/volumes?q={book_title}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if 'items' in data:
            item = data['items'][0]  # Assuming the first item is the desired book
            volume_info = item.get('volumeInfo', {})
            publication_year = volume_info.get('publishedDate')
            genre = volume_info.get('categories')
            page_count = volume_info.get('pageCount')
            publisher = volume_info.get('publisher')
            dimensions = volume_info.get('dimensions')
            # Determine whether the book is fact or fiction
            if genre is not None and 'Fiction' in genre:
                is_fiction = True
            else:
                is_fiction = False
            return publication_year, genre, page_count, publisher, dimensions, is_fiction
    return None, None, None, None, None, None

# Assuming df is your pandas DataFrame
for index, row in df.iterrows():
    book_title = row['Book']
    publication_year, genre, page_count, publisher, dimensions, is_fiction = get_book_details(book_title)
    df.loc[index, 'publication_year'] = publication_year
    df.loc[index, 'genre'] = genre
    df.loc[index, 'page_count'] = page_count
    df.loc[index, 'publisher'] = publisher
    df.loc[index, 'dimensions'] = dimensions
    df.loc[index, 'is_fiction'] = is_fiction
```

```python
df.drop(columns=['dimensions'], inplace=True)
```

```python
df.rename(columns={
    'Author': 'Author',
    'Book': 'Book',
    'Rating': '5 Star Rating',
    'Blurb': 'Blurb',
    'publication_year': 'Publication Year',
    'genre': 'Genre',
    'page_count': 'Page Count',
    'publisher': 'Publisher',
    'is_fiction': 'Is it Fiction'
}, inplace=True)
```

```python
df['Publication Year'] = pd.to_datetime(df['Publication Year']).dt.year
df['Publication Year'].fillna(0, inplace=True)
df['Publication Year'] = df['Publication Year'].astype(int)
```

```python
# Initialize empty lists for author, year of birth, and country
authors = []
birth_years = []
countries = []

# Regex pattern to extract author's name, year of birth, and country
pattern = r"^(.*?)\(born\s(\d+)\),\s(.+)$"

# Parse each entry in novelist_info
for entry in novelist_info:
    # Use regular expressions to extract author's information
    match = re.match(pattern, entry)
    if match:
        # Extract author's name, year of birth, and country
        author = match.group(1).strip()
        year_of_birth = match.group(2)
        country = match.group(3).strip()
        
        # Append to respective lists
        authors.append(author)
        birth_years.append(year_of_birth)
        countries.append(country)

# Create DataFrame
info = pd.DataFrame({
    'Author': authors,
    'Year of Birth': birth_years,
    'Country': countries
})

```

```python
import pandas as pd

# Define a list to store novelist information
novelist_info = []

# Define a flag to indicate when to start appending
start_appending = False

# Loop through each list and extract the text
for ul in lists:
    # Find all list items within the unordered list
    for li in ul.find_all('li'):
        # Check if the author is Chris Abani to start appending
        if "Chris Abani" in li.get_text(strip=True):
            start_appending = True
        # Check if the author is Norbert Zongo to stop appending
        if "Norbert Zongo" in li.get_text(strip=True):
            start_appending = False
            novelist_info.append(li.get_text(strip=True))
            break
        # Append the text to novelist_info if the flag is True
        if start_appending:
            # Extract author and country
            author_country = li.get_text(strip=True).split(",")
            author = author_country[0].strip()
            country = author_country[-1].strip()
            # Extract year of birth
            year_of_birth = None
            # Find the year of birth within parentheses
            for part in author.split():
                if part.startswith("(") and part.endswith(")"):
                    year_of_birth = part.strip("()")
                    # Remove year of birth from author
                    author = author.replace(part, "").strip()
                    break
            novelist_info.append((author, year_of_birth, country))

# Filter out tuples with more than three elements
novelist_info = [info[:3] for info in novelist_info]

# Create a DataFrame
info = pd.DataFrame(novelist_info, columns=['Author', 'Year of Birth', 'Country'])

# Display the DataFrame
info
```

```python
for index, row in info.iterrows():
    if "(" in row['Author'] and ")" in row['Author']:
        start_index = row['Author'].find("(")
        end_index = row['Author'].find(")")
        year_of_birth = row['Author'][start_index+1:end_index]
        # Update the Author column
        info.at[index, 'Author'] = row['Author'][:start_index].strip()
        # Update the Year of Birth column
        info.at[index, 'Year of Birth'] = year_of_birth
```
